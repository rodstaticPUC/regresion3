# Regresion

 Generacion de una muestra aleatoria para regresion. Principales parametros: **n_samples**, **n_feature**, sesgo (**bias**) y estado aleatorio (**random_state**).

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
# seleccion de 100 muestras
X,y = make_regression(n_samples=100, n_features=2, bias=0.2, random_state=10)
```

```python
plt.figure(figsize=(8,3))
plt.plot(X,y,'go',label='feature')
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.title('Regresion')
plt.grid(True)
```
